# 4. Data Concerns & Solutions

## Describe any expected challenges in the dataset (missing values, unstructured data, etc.)

Expected Challenges:
1) Sampling Limitations: If just a k-rows sample is used, without random sampling techniques, missing value percentages and data trends observed may not reflect the population dataset - a risk of sample bias.

2) Potential Data Inconsistency: Store Name, Address, City, Store Location, County, Category Name, Vendor Name are free-text and could suffer from minor inconsistencies like different spelling of the same category, casing, extra spaces.
For example: Store Location variable uses Longitude-Latitude data format and we can see some missing values there, which might suggest that there are inconsistencies in geocoding.

3) Missing Values: Some fields such as Store Location, Category, and Category Name contain missing entries, which can reduce data completeness and limit certain types of analysis (e.g., spatial or category-level insights).

4) Meaningless Duplicate Values: Although "Invoice/Item Number" is unique in the sample, there remains a broader risk of duplicate or redundant records across the dataset - for example, multiple entries for the same store, product, and date due to partial order submissions, system errors, or data integration issues. Duplicates, if present, could inflate sales metrics or distort performance analyses.

## Explain proposed modifications to ensure data suitability for business intelligence.


1) Randomized Full-Sample Analysis: To reduce sampling bias, ensure that analysis is conducted on either the full dataset or a properly randomized sample. Sampling should be stratified by key dimensions (e.g., date, store, product category) if full processing is not feasible due to size.

2) Standardization of Free-Text Fields: Apply consistent formatting to categorical text fields such as Store Name, Category Name, and Vendor Name. This includes trimming whitespace, enforcing consistent casing (e.g., title case), and mapping known variants to a standard label.

3) Imputation of Missing Values: For missing values in fields like Store Location or Category Name, use logical imputation strategies. For example, geocode Store Location using ZIP codes when coordinates are missing, and assign “Unknown” placeholders where category-level information is unavailable but the transaction is otherwise valid. The least option is to drop missing values that are not important.

4) Duplicate Detection: Implement logic to flag potential duplicate rows by comparing multiple fields such as Date, Store Number, Item Number, and Sale Amount. We should analyze whether duplicates actually make sense or represent valid repeated transactions (e.g., multiple sales of the same product on the same day), or whether they result from system errors or data redundancies. Based on this evaluation, either retain, consolidate, or remove duplicates to maintain data integrity.

# Creating the ORDER DIMENSION

**COMMENT:** Each row should describe a single instance of an ORDER.

```python
%%sql
DROP TABLE IF EXISTS orders;

CREATE TABLE orders (
  key SERIAL PRIMARY KEY,
  Invoice_Item_Number VARCHAR(50) NOT NULL,
  Category INTEGER,
  Category_Name VARCHAR(75),
  Item_Number INTEGER,
  Item_Description VARCHAR(100),
  Pack INTEGER,
  Bottle_Volume_mL NUMERIC
);
```

**COMMENT ON ABOVE:** Create a SURROGATE KEY (1,2,..,n) to uniquely identify each ORDER in the ORDER DIMENSION table.

**COMMENT ON BELOW:** Inserting the data from the Liquor_Sales_Transaction table into the ORDER dimension

```python
%%sql

INSERT INTO orders (
  Invoice_Item_Number,
  Category,
  Category_Name,
  Item_Number,
  Item_Description,
  Pack,
  Bottle_Volume_mL
)
SELECT
  Invoice_Item_Number,
  Category,
  Category_Name,
  Item_Number,
  Item_Description,
  Pack,
  Bottle_Volume_mL

FROM Liquor_Sales_Transactions;
```

**COMMENT ON BELOW:** Verifying the creation of the ORDER DIMENSION table

```python
%%sql
SELECT * FROM orders LIMIT 10;
```

**COMMENT ON BELOW:** Adding the new identifiers for ORDER INSTANCE (surrogate keys) back to the working FACT table (Liquor_Sales_Transaction). Specifically, by creating the column to hold the FOREIGN KEY of orders in the FACT table (named order_key in the FACT table).

```python
%%sql
ALTER TABLE Liquor_Sales_Transactions
ADD COLUMN order_key INTEGER,
ADD CONSTRAINT fk_order
  FOREIGN KEY (order_key)
  REFERENCES orders (key);
```

```python
%%sql
SELECT * FROM Liquor_Sales_Transactions LIMIT 10;
```

**COMMENT ON BELOW:** Updating the FACT table (Liquor_Sales_Transaction) so it has the same surrogate keys of the orders, where the unique id from the dimension table (orders) equals the same unique id from the fact table.

# Creating the DATE DIMENSION

**NOTE:** Date stored in format 'YYY-MM-DD'

**GOAL:** Extracting parts of dates out into their own attributes. Will be helpful for later analysis. Makes a lot of granular aggregation easier. Want to have unique dates in the DATE DIMENSION table.

**COMMENT ON BELOW:** Creating the Date Dimension

```python
%%sql
DROP TABLE IF EXISTS date;

CREATE TABLE date (
  key SERIAL PRIMARY KEY,
  Full_Date DATE,
  Year VARCHAR(4), -- Needed to change to varchar becasu
  Month_of_Year_str VARCHAR(12),
  Month_of_Year VARCHAR(4),
  Day_of_Month VARCHAR(40),
  Day_of_Week_str CHAR(9),
  Day_of_Week VARCHAR(10),
  is_weekend BOOLEAN,
  is_weekday BOOLEAN,
  Quarter_of_Year VARCHAR(10)
);
```

**COMMENT ON BELOW:** Creating ETL script to separate and extract relevant parts of date from DATE column (in facts table) into own columns to be held in the DATE DIMENSION table.

**COMMENT ON BELOW:** ETL - loading values into the relevant columns contained into the DATE DIMENSION table from the FACTS table. This will populate the DATE DIMENSION table accurately and uniquely.

```python
%%sql
INSERT INTO date (
  Full_Date,
  Year,
  Month_of_Year_str,
  Month_of_Year,
  Day_of_Month,
  Day_of_Week_str,
  Day_of_Week,
  is_weekend,
  is_weekday,
  Quarter_of_Year
)
SELECT DISTINCT
  Date as Full_Date,
  TO_CHAR(Date, 'YYYY') as Year,
  TO_CHAR(Date, 'Month') as Month_of_Year_str,
  TO_CHAR(Date, 'MM') as Month_of_Year,
  TO_CHAR(Date, 'DD') as Day_of_Month,
  TO_CHAR(Date, 'Day') as Day_of_Week_str,
  EXTRACT(DOW FROM Date) as Day_of_Week,
  CASE WHEN EXTRACT(DOW FROM Date) IN (0, 6)
    THEN true
    ELSE false
  END AS is_weekend,
  CASE WHEN EXTRACT(DOW FROM Date) NOT IN (0, 6)
    THEN true
    ELSE false
  END AS is_weekday,
  TO_CHAR(Date, 'Q') as Quarter_of_Year
FROM Liquor_Sales_Transactions;
```

**COMMENT ON ABOVE CODE:** Found (using ChatGPT) that the EXTRACT key ran on the parameter (DOW FROM DATE) will return integers 0 (designating Sunday) and 6 (to designate Saturday).

Booleans should also not be entered with quotes ('true' != true).

```python
%%sql
SELECT * FROM date LIMIT 10;
```

**COMMENT ON BELOW:** Making sure weekend/weekday was accurately processed.

```python
%%sql
SELECT DISTINCT Day_of_Week_str, Day_of_Week, is_weekend, is_weekday
FROM date
ORDER BY Day_of_Week;
```

```python
%%sql
ALTER TABLE Liquor_Sales_Transactions
ADD COLUMN date_key INTEGER,
ADD CONSTRAINT fk_date
  FOREIGN KEY (date_key)
  REFERENCES date (key);
```

## Query Development

### 1) Top preforming products and categories
- Maybe derrive a column for cost_dollars, which is simply state_bottle_cost*bottles_sold to better track top preformming products

- A potential dimension to add on would be date, where we could attatch quarter as one of the columns in the query

#### Here we look at top performing item descriptions in the dataset as a whole

```python
%%sql
SELECT
  Item_Description,
  SUM(sale_dollars) AS Total_Sales,
  COUNT(*) AS Transactions,
  ROUND(AVG(Sale_dollars),2) AS Avg_Sale,
  SUM(Sale_dollars - (Bottles_sold * State_bottle_cost)) AS Total_Profit,
ROUND(AVG(Sale_dollars) - AVG(Bottles_sold * State_bottle_cost),2) AS Average_Profit

FROM
  Liquor_Sales_Transactions, Orders

WHERE
    Liquor_Sales_Transactions.order_key = orders.key
GROUP BY
  Item_description

ORDER BY
  transactions DESC

LIMIT 10; -- lets look at the Black Velvet and Titos handmade vodka for each year
```

#### Here we look at the top performing category names in the dataset as a whole

```python
%%sql
SELECT
  Category_Name,
  SUM(sale_dollars) AS Total_Sales,
  COUNT(*) AS Transactions,
  ROUND(AVG(Sale_dollars), 2) AS Avg_Sale,
  ROUND(SUM(Sale_dollars - (Bottles_sold * State_bottle_cost)), 2) AS Total_Profit,
ROUND(AVG(Sale_dollars) - AVG(Bottles_sold * State_bottle_cost), 2) AS Average_Profit

FROM
  Liquor_Sales_Transactions, Orders

WHERE
    Liquor_Sales_Transactions.order_key = orders.key
GROUP BY
  Category_Name

ORDER BY
  Transactions DESC

LIMIT 10; -- lets look at the Black Velvet and Titos handmade vodka for each year
 --Lets look at Canadien whiskes, straight bourbon whiskeys, and whiskey liquer
```

#### Looking at top item descriptions by quarter for 'BLACK VELVET' and 'TITOS HANDMADE VODKA'

```python
%%sql
SELECT
  Item_Description,
  SUM(Sale_Dollars) AS Total_Sales,
  COUNT(*) AS Transactions,
  ROUND(AVG(Sale_Dollars), 2) AS Avg_Sale,
  quarter_of_year,
ROUND(SUM(Sale_dollars - (Bottles_sold * State_bottle_cost)), 2) AS Total_Profit,
ROUND(AVG(Sale_dollars) - AVG(Bottles_sold * State_bottle_cost),2) AS Average_Profit


FROM
  Liquor_Sales_Transactions, Orders, date
Where
    Liquor_Sales_Transactions.order_key = orders.key and
    Liquor_Sales_Transactions.date_key = date.key and

    (orders.item_description = 'BLACK VELVET' or
    orders.item_description = 'TITOS HANDMADE VODKA')

GROUP BY
  Item_Description, quarter_of_year
ORDER BY
  quarter_of_year asc
LIMIT 10;
```

#### Looking at top Category_Names by quarter for STRAIGHT BOURBON WHISKIES and CANADIAN WHISKIES

```python
%%sql
SELECT
  Category_Name,
  SUM(Sale_Dollars) AS Total_Sales,
  COUNT(*) AS Transactions,
  ROUND(AVG(Sale_Dollars), 2) AS Avg_Sale,
ROUND(SUM(Sale_dollars - (Bottles_sold * State_bottle_cost)), 2) AS Total_Profit,
ROUND(AVG(Sale_dollars) - AVG(Bottles_sold * State_bottle_cost),2) AS Average_Profit,
    quarter_of_year
FROM
  Liquor_Sales_Transactions, Orders, date
Where
    Liquor_Sales_Transactions.order_key = orders.key and
    Liquor_Sales_Transactions.date_key = date.key and

    (orders.Category_Name = 'AMERICAN VODKAS' or
    orders.Category_Name = 'CANADIAN WHISKIES')

GROUP BY
  Category_Name, quarter_of_year
ORDER BY
  quarter_of_year asc
LIMIT 10;
```

### 2) Best sales performance - counties and cities

#### First query is to get general idea as to which counties are the highest selling counties and cities
- We can see Des Moines, Cedar Rapids, and Davenport are the highest cities by transactions

```python
%%sql
SELECT
  County,
  City,
  SUM(Sale_Dollars) AS Total_Sales,
  COUNT(*) AS Number_of_Transactions,
  ROUND(AVG(Sale_Dollars), 2) AS Average_Sale_Per_Transaction,
    ROUND(AVG(Sale_dollars) - AVG(Bottles_sold * State_bottle_cost),2) AS Average_Profit

FROM
  Liquor_Sales_Transactions, store
Where
    Liquor_Sales_Transactions.store_key = store.key
GROUP BY
  County,
  City
ORDER BY
  number_of_transactions DESC;
```

#### Second query we look at those top 3 cities, and then the two selected categories of liquor (STRAIGHT BOURBON WHISKIES and Canadien Whiskey)

```python
%%sql
SELECT
  County,
  City,
  SUM(Sale_Dollars) AS Total_Sales,
  COUNT(*) AS Number_of_Transactions,
 Category_Name,
      ROUND(AVG(Sale_Dollars), 2) AS Average_Sale_Per_Transaction,
    ROUND(AVG(Sale_dollars) - AVG(Bottles_sold * State_bottle_cost),2) AS Average_Profit
FROM
  Liquor_Sales_Transactions, store, orders
Where
    Liquor_Sales_Transactions.store_key = store.key AND
    Liquor_Sales_Transactions.order_key = orders.key AND

    (orders.Category_Name = 'AMERICAN VODKAS' or
    orders.Category_Name = 'CANADIAN WHISKIES') AND

    (store.city = 'DES MOINES' or
    store.city = 'CEDAR RAPIDS' or
    store.city = 'DAVENPORT')

GROUP BY
  County,
  City,
  Category_Name
ORDER BY
  city DESC;
```

### 3) Weekend vs Weekday Sales Performance (Categories)

```python
%%sql
Select is_weekend,
    sum(sale_dollars) as Total_Summ,
    ROUND(AVG(sale_dollars), 2) as avg_sale,
    COUNT(*) AS Transactions,
    ROUND(AVG(Sale_Dollars), 2) AS Average_Sale_Per_Transaction,
    ROUND(AVG(Sale_dollars) - AVG(Bottles_sold * State_bottle_cost),2) AS Average_Profit,
    Category_Name
From Liquor_Sales_Transactions, date, orders --facts and 2 dimensions
WHERE Liquor_Sales_Transactions.date_key = date.key AND
    Liquor_Sales_Transactions.order_key = orders.key AND

    (orders.Category_Name = 'AMERICAN VODKAS' or
    orders.Category_Name = 'CANADIAN WHISKIES')

GROUP BY is_weekend,Category_Name;
```

### 4) Best performing day of the week

#### Best preforming day of week overall

### 3) VISUALIZATION: Weekend vs Weekday Sales Performance (Products & Categories)

```python
result = %sql SELECT is_weekend, \
    SUM(sale_dollars) AS total_summ, \
    COUNT(*) AS transactions, \
    ROUND(AVG(Sale_Dollars), 2) AS average_sale_per_transaction, \
    ROUND(AVG(Sale_Dollars) - AVG(Bottles_sold * State_bottle_cost), 2) AS average_profit, \
    category_name \
FROM Liquor_Sales_Transactions, date, orders \
WHERE Liquor_Sales_Transactions.date_key = date.key AND \
      Liquor_Sales_Transactions.order_key = orders.key AND \
      (orders.category_name = 'AMERICAN VODKAS' OR \
       orders.category_name = 'CANADIAN WHISKIES') \
GROUP BY is_weekend, category_name;
```

```python
# import libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Load and clean data
df = result.DataFrame()
df['category_name'] = df['category_name'].str.strip()
df['average_profit'] = pd.to_numeric(df['average_profit'], errors='coerce')
df['transactions'] = pd.to_numeric(df['transactions'], errors='coerce')
df['is_weekend'] = df['is_weekend'].astype(bool)
df.dropna(subset=['average_profit', 'transactions'], inplace=True)

colors = ["#73B3D1", "#FCDD88"]

# Set font
plt.rcParams["font.family"] = "Roboto"

# Prepare figure
fig, axes = plt.subplots(1, 2, figsize=(29, 12), sharey=False)

# First plot - Transactions subplot
pivot_ax1 = df.pivot(index='is_weekend', columns='category_name', values='transactions')
plot1 = pivot_ax1.plot(kind='bar', stacked=True, ax=axes[0], color = colors)

# Design
# Title
axes[0].set_title("Total Number of Transactions by Category", fontsize=30, fontweight='bold')

# X-axis
axes[0].set_xlabel("Is Weekend", fontsize=26)
axes[0].tick_params(axis='x', labelrotation=0, labelsize=22)
axes[0].set_xticklabels(["Weekdays", "Weekends"], fontsize=22)

# Y-axis
axes[0].set_ylabel("Number of Transactions", fontsize=26)
axes[0].tick_params(axis='y', labelsize=22)

# Legend
axes[0].legend(title='Category', fontsize=22, title_fontsize=22)

# Labels
for container_idx, container in enumerate(plot1.containers):
    for bar_idx, bar in enumerate(container):
        # Match index from the DataFrame
        is_weekend = pivot_ax1.index[bar_idx]
        if not is_weekend:  # Weekdays only
            axes[0].text(bar.get_x() + bar.get_width()/2,
                         bar.get_y() + bar.get_height()/2,
                         f"{int(bar.get_height()):,}",
                         ha='center', va='center',
                         fontsize=26, color='black')

# Plot 2 - Average Profit
pivot_ax2 = df.pivot(index='is_weekend', columns='category_name', values='average_profit')
plot2 = pivot_ax2.plot(kind='bar', stacked=True, ax=axes[1], color = colors)

# Design
# Title
axes[1].set_title("Average Profit by Category", fontsize=30, fontweight='bold')

# X-axis
axes[1].set_xlabel("Is Weekend", fontsize=26)
axes[1].tick_params(axis='x', labelrotation=0, labelsize=22)

# Y-axis
axes[1].set_ylabel("Average Profit ($)", fontsize=26)
axes[1].tick_params(axis='y', labelsize=22)
axes[1].set_xticklabels(["Weekdays", "Weekends"], fontsize=22)


# Legend
axes[1].legend(title='Category', fontsize=22, title_fontsize=22)

# Labels
for container in plot2.containers:
    axes[1].bar_label(container, labels=[f"${int(v)}" for v in container.datavalues],
                      fontsize=26, label_type='center', color='black')

# Final layout and save
sns.despine(top=True)
plt.tight_layout()
plt.savefig("sales_by_day_type_and_category.png", dpi=300, bbox_inches='tight')
plt.show()
```

#### Interpretation/Business Insight for WEEKEND VS WEEKDAY SALES PERFORMANCE (PRODUCTS & CATEGORIES)

Across time in this dataset, alcohol transactions through vendors, the Iowa State Government, and retailers occur much more frequently on Weekdays for American Vodkas and Canadian Whiskeys (relative to Weekends). Although American Vodkas sell at a higher volume compared to Canadian Whiskies, Canadian Whiskies offer a higher Average Profit per transaction, both on Weekdays and Weekends. It is notable that the difference in Profit between the two, on average, increases quite substantially on weekends (44 dolalrs higher average profit for Canadian Whiskies) vs weekdays (11 dollars higher average profit for Canadian Whiskies). Therefore, if the Iowa State Government is going to sell liquor on the weekend, they should prioritize selling Canadian Whiskies as opposed to American Vodkas.

American Vodkas highest performing day, in terms of Average Profit, would only be higher than the two worst performing days of sales, by the same metric, of Canadian Whiskies. If the Iowa State Government can keep the returns in average profit high for Canadian Whiskies Saturday sales, they should emphasize trying to increase their sales of Canadian Whiskies on Saturday.
